```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
import os
```

```python
numero_de_archivos=len(os.listdir("resultados_eigen_correct"))

probabilidades=np.zeros(numero_de_archivos)
gap=np.zeros(numero_de_archivos)
cont=0
for resultado in os.listdir("resultados_eigen_correct"):
    data=np.genfromtxt("resultados_eigen_correct/"+resultado,delimiter=" ")
    probabilidades[cont]=data[0]
    gap[cont]=data[1]
    cont+=1
```

```python
M=probabilidades.size
cuentas,bins=np.histogram(probabilidades,bins=100)

freq=cuentas/M

error=np.sqrt(freq*(1-freq)/M)/(bins[1]-bins[0])
```

```python
plt.figure(1,figsize=(10,6))

plt.title("Histogram of the success probability",fontsize=20)

counts,bins,axes=plt.hist(probabilidades,bins=100,histtype="step",color="#68f0ea",density=True)
plt.ylabel("Frequency",fontsize=16)
plt.xlabel("Probability",fontsize=16)
plt.xlim(0,1)

plt.errorbar(bins[:-1]+(bins[1]-bins[0])/2,counts,yerr=error,linewidth=0.0,elinewidth=1.0,capsize=2,ecolor="red")
plt.savefig("histogram_success_probability_eigen2_correct.png",format="png",bbox_inches="tight")
```

```python
plt.figure(1,figsize=(10,6))

plt.title(r"Probability as a function of $\Delta\lambda$",fontsize=20)
plt.plot(gap,probabilidades,marker="x",markersize=4,linewidth=0.0)
plt.xlabel(r"$\lambda_{max}-\lambda_{min}$",fontsize=15)
plt.ylabel(r"Probability",fontsize=15)
plt.xlim(0.0,gap.max())
plt.ylim(0.0,1.0)
plt.savefig("gap_vs_prob_correct.png",format="png",bbox_inches="tight")
```
